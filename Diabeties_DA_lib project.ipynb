```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset
df = pd.read_csv(r"C:\Users\User\Downloads\Dataset of Diabetes .csv")
df
```

```python
# Display first 5 rows
print(df.head())
```

```python
# Dataset Information
print("\nDataset Info:")
print(df.info())
```

```python
# Check Missing Values
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
# Statistical Summary
print("\nStatistical Summary:")
print(df.describe())
```

```python
# NumPy Analysis

average_bmi = np.mean(df["BMI"])
average_hba1c = np.mean(df["HbA1c"])

print("\nAverage BMI:", average_bmi)
print("Average HbA1c:", average_hba1c)
```

```python
# Diabetes Class Count

print("\nDiabetes Class Distribution:")
print(df["CLASS"].value_counts())
```

```python
# Visualization 1 - Diabetes Class Distribution

plt.figure(figsize=(6,4))
sns.countplot(x="CLASS", data=df)

plt.title("Diabetes Class Distribution")
plt.xlabel("Class")
plt.ylabel("Count")

plt.show()
```

```python
# Visualization 2 - BMI Distribution

plt.figure(figsize=(8,5))
sns.histplot(df["BMI"], bins=20, kde=True)

plt.title("BMI Distribution")
plt.xlabel("BMI")

plt.show()
```

```python
# Visualization 3 - HbA1c vs BMI

plt.figure(figsize=(8,5))
sns.scatterplot(x="BMI", y="HbA1c", hue="CLASS", data=df)

plt.title("BMI vs HbA1c")
plt.xlabel("BMI")
plt.ylabel("HbA1c")

plt.show()
```

```python
# Visualization 4 - Correlation Heatmap

numeric_df = df.select_dtypes(include=np.number)

plt.figure(figsize=(10,6))
sns.heatmap(numeric_df.corr(), annot=True)

plt.title("Correlation Heatmap")

plt.show()
```

```python
# Gender-wise Average HbA1c

gender_hba1c = df.groupby("Gender")["HbA1c"].mean()

print("\nAverage HbA1c by Gender:")
print(gender_hba1c)
```
